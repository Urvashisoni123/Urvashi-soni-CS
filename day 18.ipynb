#### Graph Functions:
1. plot                    - Simple Line Plot
2. plot.area               - Filled version of Line Plot
3. plot.bar / plot.barh    - Bar Graph
4. plot.box                - box Plot
5. plot.density            - Density Line (Smooth Curved)
6. plot.hexbin             - Hexagon Version of scatter with density 
7. plot.hist               - Histogram
8. plot.kde                - kernel Density Estimator (density graph)
9. plot.pie                - Pie chart of categorical data
10. plot scatter           - Scatter graph

```python
fig = plt.figure(figsize=(20,20))
plt.subplot(4,3,1);
plt.subplot(4,3,2);df2.plot();plt.title('Line Plot')
plt.subplot(4,3,3);df2.plot.area();plt.title('Area Plot')
plt.subplot(4,3,4);df2.plot.bar();plt.title('Bar Plot')
plt.subplot(4,3,5);df2.plot.density();plt.title('Density Plot')
plt.subplot(4,3,6);df2.plot.hexbin(x='a',y='b',C='c',gridsize=6);plt.title('Hexbin Plot')
plt.subplot(4,3,7);df2.plot.hist();plt.title('Hist Plot')
plt.subplot(4,3,8);df2.plot.kde();plt.title('Kde Plot')
plt.subplot(4,3,9);df2.plot.scatter(x='a',y='b');plt.title('Scatter Plot')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe(include='all')
```

```python
tips.value_counts('day')
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot('day', data=tips)
```

```python
sns.countplot(x='day',data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips, hue='time',estimator=np.mean)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips ,hue='sex')
```
